Colored Images

```python
from skimage import data # getting the image
import matplotlib.pyplot as plt

cat = data.chelsea()
plt.imshow(cat)
```

```python
# It just a ndarray
cat.shape
# The first two dimension are rows and coloumn and the third dimension is RGB colour.
```

```python
red_cat = cat.copy()
red_cat[:,:,1] = 0
red_cat[:,:,2] = 0
plt.imshow(red_cat)
```

```python
red_cat = cat.copy()
red_cat[:,:,0] = 0
red_cat[:,:,2] = 0
plt.imshow(red_cat)
```

```python
red_cat = cat.copy()
red_cat[:,:,1] = 0
red_cat[:,:,0] = 0
plt.imshow(red_cat)
```

```python
# Cropping the image 
remove_last_column = cat[:,:-1,]
print(remove_last_column.shape)
```

```python
eye_cat = remove_last_column[80:150,120:220]
plt.imshow(eye_cat)
```

```python
# Transpose the image 
plt.imshow(cat.transpose((1,0,2)))
```

```python
remove_last_column[80:170,120:220,0] = 0
eye_cate = remove_last_column.copy()
eye_cate[80:170,280:380,1] = 0
plt.imshow(eye_cate) 
```

```python
copy_cat = cat.copy()
copy_cat[copy_cat[:,:,0] > 160] = [0,255,0]
plt.imshow(copy_cat)
```
