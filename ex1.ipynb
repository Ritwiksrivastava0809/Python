TEMPRATURE DATA IN NEW YORK

```python
import numpy as np

record_low_list = [-28,-22,-21,9,25,35,40,34,24,16,-11,-22]
record_low_array = np.array(record_low_list)
list1 = []
for i in record_low_array : 
    c = (5/9)*(i - 32)
    list1.append(c)
record_low_celcius = np.array(list1)    
print("record_low_celcius : ",record_low_celcius)
print()
print("Record low in May : ",record_low_celcius[4])
print()
print("Record low in September : ",record_low_celcius[8])
print()
record_low_celcius_spring = np.array(record_low_celcius[2:5])
print(record_low_celcius_spring)
print()
sorted_celcius = np.array(np.sort(record_low_celcius))
print("sorted_celcius : ",sorted_celcius)
print()
print("Max temprature in dataset is : np.",max(record_low_celcius))
print()
print("Min temprature in dataset is : ",np.min(record_low_celcius))
print()
print("Max temprature is in month : ",np.argmax(record_low_celcius)+1)
print()
print("Min temprature is in month : ",np.argmin(record_low_celcius)+1)
print()
print("The average low temprature in whole year is: ",np.mean(record_low_celcius))
print()
```

```python
months = np.array(["JAN","FEB","MAR","APRIL","MAY","JUNE","JULY","AUG","SEPT","OCT","NOV","DEC"])
import matplotlib.pyplot as plt
plt.bar(months,record_low_celcius)
plt.xlabel("Months")
plt.ylabel("Temprature")
plt.title("NEW YORK TEMPRATURE DATA")
```

```python
amplitude = (np.max(record_low_celcius) - np.min(record_low_celcius))/2
months = np.arange(1,13)
vector = (2*np.pi*(months - np.argmax(record_low_celcius)))/12
apx = (amplitude * np.cos(vector) + np.mean(record_low_celcius))
y_values = record_low_celcius - apx

plt.bar(months,y_values)
plt.xlabel("Months")
plt.ylabel("Temprature Diff")
plt.title("NEW YORK TEMPRATURE DATA")
```

```python
x_values = np.linspace(0,12,100)
apx1 = (amplitude * np.cos((2*np.pi*(x_values - np.argmax(record_low_celcius)))/12) + np.mean(record_low_celcius))
plt.plot(x_values,apx1)
```
