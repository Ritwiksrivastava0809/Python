MATHEMATICAL COJECTURE

```python
import numpy as np
import matplotlib.pyplot as plt

no = 50

y_values = []
while (no != 1) :
    if no % 2 != 0 :
        no = (no * 3) +1
        y_values.append(no)
    else :
        no = no / 2
        y_values.append(no)    
             

plt.figure(figsize=(16,10), dpi= 80)#zoom in zoom out
print(y_values)
print(len(y_values))
x_values = np.arange(len(y_values))
plt.scatter(x_values,y_values)
plt.plot(x_values,y_values,color = 'red')
# plt.xticks(np.arange(0,no).tolist())#scaling
```
