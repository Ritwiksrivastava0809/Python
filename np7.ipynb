1D DISCREAT FOURIER TRANSFORMATION

```python
from numpy import fft
import matplotlib.pyplot as plt
```

```python
num_samples = 400
sample_freq = 1/200 
time = np.linspace(0,sample_freq*num_samples,num_samples)
```

```python
simple_signal = np.sin(20*np.pi*time)
```

```python
plt.plot(time,simple_signal)
```

```python
f_f=fft.rfft(simple_signal).shape
```
